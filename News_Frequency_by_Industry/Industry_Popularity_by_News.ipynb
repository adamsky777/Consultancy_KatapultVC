```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
inp = pd.read_excel("Industry_popularity_by_news.xlsx")
```

```python
# set color to custom sample:
custom_blue = (11/255, 27/255, 86/255)
# Plotting
plt.figure(figsize=(10, 6))
plt.bar(inp['PrimaryIndustrySector'], inp['FrequencySum'], color=custom_blue) # use blue or custom blue
plt.xlabel('PrimaryIndustrySector')
plt.ylabel('Frequency')
plt.title('News Frequency by Industry')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('News_frequency_by_Industry.png')
plt.savefig('News_frequency_by_Industry.svg', format='svg')

plt.show()
```
